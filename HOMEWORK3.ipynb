Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 

```python
a = int(input())
sum = 0
for i in range(1, a):
    if a % i == 0:
        sum += i
if sum == a:
    print('Perfect number')
else:
    print('Not perfect number')
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập.

```python
N = int(input())
maxnumber = N #Gán biến 
minnumber = N #Gán biến 
while True:
    if N == -1: #Nhập đến -1 thì dừng
        break
    if N > maxnumber: 
        maxnumber = N
    if N < minnumber:
        minnumber = N
    N = int(input()) #Đọc tiếp số tiếp theo
print(maxnumber, minnumber)


    
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input())
a = 1
b = 1
while (a + b) <= A:
    fnext = a + b #Phải thêm 1 biến tạm thời vào
    a = b
    b = fnext
print(b)
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố

```python
n = int(input())
dem = 0 #Biến đếm
for i in str(n)[::1]: #Phân tách n bằng string slicing
    lasnt = True #Gán tất cả các số trong n đều là số nguyên tố
    if int(i) <= 1: #Loại trừ các số 0, 1
        lasnt = False #Nếu thỏa mãn điều kiện này => Không phải là số nguyên tố
    for j in range(2, int((int(i)**0.5) + 1)): #Với số nguyên tố ta chỉ cần kiểm tra i có chia hết các số nào trong khoảng từ 2 đến căn của i + 1
        if int(i) % j == 0: #Nếu thỏa mãn điều kiện này => Không phải là số nguyên tố
            lasnt = False  
    if lasnt == True: #Loại trừ 2 trường hợp trên thì các số còn lại là số nguyên tố
         dem += 1 #Biến đếm được + 1 với mỗi giá trị i là số nguyên tố
print(dem)
```

W4A25: Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
n = int(input())
maxnumber = n
minnumber = n
while True:
    if n > maxnumber:
        maxnumber = n
    if n < minnumber:
        minnumber = n
    n = int(input())
    if n == -1:
        break
print(maxnumber, minnumber)
```
