d. 

```python
a, b, c, d = map(int, input().split())
if b%a == 0:
    r = b//a
    if ((b*r) == c) and ((c*r) == d):
        print('Cấp số nhân với công bội:', r)
    else:
        print('No')
else:
    print('No')
```
