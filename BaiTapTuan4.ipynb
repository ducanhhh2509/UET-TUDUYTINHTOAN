W4A8:Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
def reverse(n):
    N = int(str(n)[::-1])
    return N
dem = 0
n = int(input())
while n != reverse(n):
    n += reverse(n)
    dem += 1
    if n == reverse(n):
        break
print(f'Số đối xứng thu được là {n} và số bước: {dem}')
```

Ví dụ 220 và 284 là một cặp số thân thiết, 220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 110. Tổng của 11 ước này cho chúng ta 284, 284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng vừa đúng bằng 220
Yêu cầu đề bài: Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false" 

```python
n1, n2 = map(int,input().split())
if n1 > 0 and n2 > 0:
    tong1 = 0
    tong2 = 0
    for i in range(1, n1):
        if n1 % i == 0:
            tong1 += i
    for j in range(1, n2):
        if n2 % j == 0:
            tong2 += j
    if (n1 == tong2) and (n2 == tong1):
        print(f'true')
    else:
        print(f'false')
else:
    print(f'Hai số {n1} và {n2} đã nhập không thỏa mãn yêu cầu đề bài')
```

W4A14: Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
tìm ước chung lớn nhất của hai số m và n. 

```python
def GCD(a, b):
    if a > b:
        for i in range(1, b+1):
            if b % i == 0:
                if a % i == 0:
                    uocmax = i
        return uocmax
    elif a < b:
        for j in range(1, a+1):
            if a % j == 0:
                if b % j == 0:
                    uocmax = j
        return uocmax
    else:
        return a
m, n = map(int, input().split())
print(GCD(m, n))
    
```
