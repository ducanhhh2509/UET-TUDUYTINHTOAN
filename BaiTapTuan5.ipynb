W5A1: Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) ->

W5A2: Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]: 

```python
def swap(a, b):
    a, b = b, a
    return f'a = {a}, b = {b}'
a, b = map(int, input().split())
print(swap(a, b))
```

W5A3: Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def kiemtrasonguyento(a):
    lasnt = True
    for i in range(1, a+1):
        if a % i == 0 and i != 1 and i != a:
            lasnt = False
    if a < 2:
        lasnt = False
    if lasnt == True:
        return f'{a} là số nguyên tố'
    if lasnt == False:
        return f'{a} không là số nguyên tố'
a = int(input())
print(kiemtrasonguyento(a))
```

W5A4: VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def sohoanhao(a):
    tong = 0
    for i in range(1, a):
        if a % i == 0:
            tong += i
    if a == tong:
        return f'{a} là số hoàn hảo'
    elif a == 1:
        return f'{a} không là số hoàn hảo'
    else:
        return f'{a} không là số hoàn hảo'
a = int(input())
print(sohoanhao(a))
```

W5A5: Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 


```python
def timvitri(danhsach, k):
    for i in range(len(danhsach)):
        if danhsach[i] == k:
            return i
    return -1
```

W5A6: Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n

```python
def giaithua(n):
    tich = 1
    for i in range(1, n+1):
        tich *= i
    return tich
n = int(input())
print(giaithua(n))
```

W5A8: Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. 

```python
def Bin(n):
    ans = ''
    while n > 0:
        du = n % 2
        ans += str(du) 
        n //= 2
    return ans
def hamming_distance(a, b):
    s1 = Bin(a)
    s2 = Bin(b)
    if len(s1) > len(s2):
        while len(s1) > len(s2):
            s2 += '0'
    if len(s1) < len(s2):
        while len(s1) < len(s2):
            s1 += '0'
    an = 0
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            an += 1
    return an
print(hamming_distance(100, 36))
```

W5A12: Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|. 


```python
def  divide_with_remainder(dividend, divisor):
    q = dividend // divisor
    r = dividend % divisor
    return q, r
a, b = map(int, input().split())
print(divide_with_remainder(a, b))
```

W5A14: Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op.

```python
def compute(a, b, op):
    if op == '+':
        return f'{a} {op} {b} = {a+b}'
    if op == '-':
        return f'{a} {op} {b} = {a-b}'
    if op == 'x':
        return f'{a} {op} {b} = {a*b}'
    if op == ':':
        return f'{a} {op} {b} = {a/b}'
    if op != '+' or op != '-' or op != 'x' or op != ':':
        return f'phép tính không hợp lệ'
a, b = map(int, input().split())
op = input()
print(compute(a, b, op))
```

W5A15: Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def  is_right_triangle(a, b, c):
    if a > 0 and b > 0 and  c > 0 and a + b > c and a + c > b and b + c > a:
        if a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
            return True
        else:
            return False
    else:
        return False
```
