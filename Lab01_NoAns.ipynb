### Problem 6
#### Write a function that computes the area of a triangle with side lengths a, b, c (use Heron's formula). Raise ValueError if the sides cannot form a triangle.

```python
def dientich(a,b,c):
    if a>0 and b>0 and c>0 and a+b>c and a+c>b and b+c>a:
       p=(a+b+c)/2
       s=(p*(p-a)*(p-b)*(p-c))**0.5
       return s
    else:
        raise ValueError('Đây không phải là 1 tam giác')
a,b,c=map(float,input('Nhập 3 cạnh của tam giác:').split())
print(dientich(a,b,c))
```
